```python
all_files = os.listdir(data_dir) #os.listdir() is a method that returns a list containing the names of the entries in the directory given
# initilisation
target_file_name = ''
year = '2020_'

for file_name in all_files:
    # check if this file is for 2020
    is_year = file_name.startswith(year) # YOUR CODE HERE
    # if yes, we check if the file name is for the country that you select
    if is_year:
        # check if the country_code is in the file_name
        is_country = file_name.startswith(year + country_code) # YOUR CODE HERE
        if is_country:
            # found the filefile_name.startswith(year), save it to 'target_file_name'
            target_file_name = file_name  # YOUR CODE HERE
            # we stop looking by breaking out of the for loop
            break
if not target_file_name:
    print('File not found. Check your country code (or select a different one)!')
else:
    print('Found file name is: ' + target_file_name)
    # get the path to the file
    file_path = './Region_Mobility_Report_CSVs/2020_GR_Region_Mobility_Report/' # YOUR CODE HERE
    print('Path to the file is: ' + file_path)
```

#### 1. There are multiple activities in the mobility report such as workplaces, parks, transit stations, etc. For each of the activities, what are the average percent changes from baselines for your selected country?

```python
# YOUR CODE HERE
workplace_mean = df['workplaces_percent_change_from_baseline'].mean() #calculate the mean 
workplace =float("{0:.3f}".format(workplace_mean)) # keep 3 decimals in the presented result

parks_mean = df['parks_percent_change_from_baseline'].mean()
parks =float("{0:.3f}".format(parks_mean))

trans_station_mean = df['transit_stations_percent_change_from_baseline'].mean()
transit =float("{0:.3f}".format(trans_station_mean))

print('workplace_mean=', workplace)
print('parks_mean=', parks ) 
print('transit_station_mean=', transit)
```

#### 5. We're going to look at the changes of this province over time. As you can see, the column date is in string format. Convert this str date column data str to datetime format and append as an extra column called 'date2' to the dataframe.  
Hint: <https://pandas.pydata.org/docs/reference/api/pandas.to_datetime.html>

```python
# YOUR CODE HERE
date2 = pd.to_datetime(df_new['date'])

df_new.insert(loc=8, column ='date2', value=date2)

df_new
```

#### 6. Sort the data in ascending order of time and save it as 'df_province_largest'

```python
# YOUR CODE HERE
df_province_largest= df_new.sort_values(by='date2')

df_province_largest
```

```python
# Execute the cell with the following command to visualize the results
df_province_largest.plot('date2', 'workplaces_percent_change_from_baseline')

df_province_largest.resample('7D', on='date2').sum()['workplaces_percent_change_from_baseline'].plot()
```
